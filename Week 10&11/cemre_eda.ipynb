```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
customer_data = pd.read_csv("Week 9/df_preprocessed.csv")
```

```python
customer_data.info()
```

```python
customer_data.isna().sum()
```

```python
customer_data = customer_data[customer_data['cust_residence'].notna()]
```

```python
customer_data.isna().sum()
```

```python
customer_data.nunique() 
```

```python
customer_data = customer_data.astype({'emp_index':'category', 'cust_residence': 'category', 'cust_gender':'category', 'cust_rel_time':'category', 'name_province':'category'})
customer_data = customer_data.rename(columns = {'long_term_deposit':'medium_term_deposit'}) 
customer_data = customer_data.rename(columns = {'ind_dela_fin_ult1':'long_term_deposit'}) 
```

```python
customer_data.info()
```

```python
customer_data['age'].describe()
```

```python
#age distribution
sns.set(rc={'figure.figsize':(20,10)})
sns.countplot(x = customer_data['age'], data = customer_data)
```

```python
bins = [18, 25, 60, 85]
labels = ['Young', 'Adult', 'Senior']
customer_data['Age_Range'] = pd.cut(customer_data['age'], bins=bins, labels=labels, right=False)
```

```python
customer_data.head()
```

```python
sns.catplot(data = customer_data, x = 'Age_Range', kind='count', aspect=2)
```

```python
#gender distribution
customer_data.cust_gender = customer_data.cust_gender.replace({'H': 'Male', 'V':'Female'})
customer_data_gender = customer_data['cust_gender'].value_counts()
customer_data_gender.plot(kind='pie', autopct='%1.1f%%').set_title("Gender distribution")
```

```python
#gender distribution by age
sns.catplot(data = customer_data, x = 'Age_Range', kind='count', hue = 'cust_gender', aspect=2)
```

```python
customer_data_gender = customer_data.loc[:, ['cust_gender', 'activity_index']].join(customer_data.loc[:, "saving_acc": "direct_debit"])
customer_data_gender = customer_data_gender.groupby(['cust_gender', 'activity_index']).sum()
customer_data_gender = customer_data_gender.T
customer_data_gender.head()
```

```python
customer_data_gender.plot(kind='barh', stacked=True)
plt.title('Popularity of Products by Gender and Activity Index', fontsize=16) 
plt.xlabel('Number of customers', fontsize=15) 
plt.ylabel('Product names', fontsize=15) 
plt.legend(["Gender:M, Activity Index:0", "Gender:M, Activity Index:1", "Gender:F, Activity Index:0", "Gender:F, Activity Index:1"]) 
```

```python
# 4 major channels
```

```python
customer_data_channel = customer_data.loc[:, ['channel_to_join']].join(customer_data.loc[:, "saving_acc": "direct_debit"])
customer_data_channel = customer_data_channel.loc[customer_data_channel['channel_to_join'].isin(["KAT", "KFC", "KHE", "KFA"])]
customer_data_channel = customer_data_channel.groupby("channel_to_join").sum()
customer_data_channel = customer_data_channel.T
customer_data_channel
```

```python
customer_data_channel.plot(kind='barh', stacked=True)
plt.title('Channels Used to Join', fontsize=16) 
plt.xlabel('Number of customers', fontsize=15) 
plt.ylabel('Product names', fontsize=15) 
```

```python
customer_data_income = customer_data.loc[:, ['name_province']].join(customer_data['household_income'])
customer_data_income = customer_data_income.groupby('name_province').mean()
customer_data_income.sort_values(by = "household_income", inplace=True)
customer_data_income.reset_index(inplace=True)
customer_data_income.head()
```

```python
sns.factorplot(data = customer_data_income, x="name_province", y= "household_income", size = 10)
plt.xticks(rotation=90)
plt.ylabel("Median Income")
plt.xlabel("City")
plt.title("Income Distribution by City")
```
