```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# import plotly.express as px
# import plotly.io as pio
# template = 'simple_white'

import seaborn as sns
sns.set()

from sklearn import metrics

import warnings
warnings.filterwarnings('ignore')

plt.rcParams['axes.labelsize'] = 15
plt.rcParams['axes.titlesize'] = 20

seed = 42
```

```python
df = pd.read_csv("C:/Users/joean/Desktop/Data Science/Personal Projects/df_preprocessed.csv")
df
```

```python
df.info()
```

```python
import pickle
col_transformer = pickle.load(open('col_transformer.pkl', 'rb'))
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import QuantileTransformer
from sklearn.decomposition import PCA

transformer = make_pipeline(
    (col_transformer), 
    (QuantileTransformer(output_distribution = "normal", random_state = seed))
)

arr = transformer.fit_transform(df)
```

```python
%%time
from tensorflow import keras

encoder = keras.models.Sequential([
    keras.layers.Dense(40, input_shape=[arr.shape[1]], activation='relu', activity_regularizer = "l1"),
    keras.layers.Dense(25, activity_regularizer = "l1", activation='relu'),
    keras.layers.Dense(3, activation='relu'),
])

decoder = keras.models.Sequential([
    keras.layers.Dense(25, input_shape=[3], activation='relu', activity_regularizer = "l1"),
    keras.layers.Dense(40, activity_regularizer = "l1", activation='relu'),
    keras.layers.Dense(arr.shape[1], activation='relu'),
])

autoencoder = keras.models.Sequential([encoder, decoder])
autoencoder.compile(optimizer = "adam", loss = "mean_squared_error")
autoencoder.summary()


history = autoencoder.fit(arr,arr, epochs=10, batch_size = 50, verbose=1)

reduced_arr = encoder.predict(arr)

reduced_df = pd.DataFrame({
    "dim1" : reduced_arr[:,0],
    "dim2" : reduced_arr[:,1],
    "dim3" : reduced_arr[:,2],
})

fig = plt.figure(figsize=(12, 7), dpi=80, facecolor='w');
ax = plt.axes(projection="3d");
ax.scatter3D(reduced_df['dim1'], reduced_df['dim2'],reduced_df['dim3'], cmap='Spectral', edgecolor='k', 
            alpha = 0.5);
```
