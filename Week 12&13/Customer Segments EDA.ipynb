```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns
sns.set()

from sklearn import metrics

import warnings
warnings.filterwarnings('ignore')

plt.rcParams['axes.labelsize'] = 15
plt.rcParams['axes.titlesize'] = 20

seed = 42
```

```python
df = pd.read_csv("C:/Users/joean/Desktop/Data Science/Personal Projects/df_with_clusters.csv")
df
```

```python
ax = sns.countplot(df['customer_segments']);
# ax.bar_label(ax.containers[0])

for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()), 
       ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
    
ax.set_title("Total Customers for Each Clusters");
```

Cluster 0 has the highest number of customers. The rest of the clusters have roughly the same amount of customers.

```python
pd.crosstab(index = df['customer_segments'], columns = df["cust_gender"], normalize = "index").\
plot(kind = "barh", stacked = True, figsize = (7,6), title = "Proportion of Genders per Cluster");
```

Interestingly, Cluster 0 and 2 are made up of only Customer with gender "V" whereas for clusters 1 and 3, it is "H".

```python
plt.figure(figsize = (8,6))
ax = sns.violinplot(y = "age", x = "customer_segments", data = df);

medians = df.groupby(['customer_segments'])['age'].median()
vertical_offset = df['age'].median() * 0.05# offset from median for display

for xtick in ax.get_xticks():
    ax.text(xtick, medians[xtick] + vertical_offset, medians[xtick], 
            horizontalalignment = 'center', size='x-large', color='black', weight='bold')
ax.set_title("Age distribution for each clusters");
```

Cluster 3 has the youngest median age of 26, followed by cluster 2 (median age of 36). Customer segment 0 and 1 have senior customers.

```python
pd.crosstab(index = df['customer_segments'], columns = df["cust_rel_time"], normalize = "index").\
plot(kind = "barh", stacked = True, figsize = (10,6), 
     title = "Proportion of Customer relation type at the beginning of the month per Cluster");
```

Clusters 1 and 0 consists of only Active customers, whereas cluster's 2 and 3 are entirely Inactive customers.

```python
pd.crosstab(index = df['name_province'], columns = df["customer_segments"], normalize = "index").\
sort_values(by = [0,1,2,3]).\
plot(kind = "barh", stacked = True, figsize = (10,10), 
     title = "Proportion of customer segments by province");

plt.xlabel("Proportion of Customer Segments");

plt.legend(bbox_to_anchor = (0.7, -0.08), ncol = 4);
```

```python
pd.crosstab(index = df['customer_segments'], columns = df["activity_index"], normalize = "index").\
plot(kind = "barh", stacked = True, figsize = (8,6), 
     title = "Proportion of customer segments by Activity Index");

plt.legend(bbox_to_anchor = (0.65, -0.08), ncol = 2);
```

Majority of Customers in segment 0 and 1 are active, whereas the opposite is observed for customers in other clusters.

```python
temp_df = df.groupby('customer_segments').agg({'household_income' : 'median'}).\
sort_values('household_income').reset_index()

plt.figure(figsize = (8, 6));
ax = sns.barplot(y = 'household_income', x = 'customer_segments', data = temp_df, palette = "viridis");

for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x() + p.get_width() / 2., p.get_height()), 
       ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
    
    plt.ylabel("Median Household Income");
plt.xlabel("Customer Segments");
```

Customers in Cluster 0 has the highest median salary and customers from cluster 3 has the lowest median salary.

```python
plt.figure(figsize = (8,6))
ax = sns.boxplot(y = "total_accounts", x = "customer_segments", data = df);

medians = df.groupby(['customer_segments'])['total_accounts'].median()
vertical_offset = df['total_accounts'].median() * 0.05# offset from median for display

for xtick in ax.get_xticks():
    ax.text(xtick, medians[xtick] + vertical_offset, medians[xtick], 
            horizontalalignment = 'center', size='x-large', color='black', weight='bold')
ax.set_title("Number of Products for each clusters");
```

Customers from segment 0 and 1 in general tend to hold more accounts compared to customers from 2 and 3.

```python
prod_cols = ['saving_acc', 'guarantees', 'current_acc', 'derivada_acc', 
             'payroll_acc', 'junior_acc', 'mass_particular_acc', 'particular_acc',
             'particular_plus_acc', 'short_term_deposit', 'medium_term_deposits', 'long_term_deposits', 
             'e_account', 'funds', 'mortgage', 'pensions1', 
             'loans', 'taxes', 'credit_card', 'securities',
             'home_account', 'payroll', 'pensions2', 'direct_debit']

temp_df = df.groupby('customer_segments').agg({col : 'sum' for col in prod_cols}).T

temp_df.div(temp_df.sum(axis = 1), axis = 0).\
sort_values(by = [0,1,2,3]).\
plot(kind = "barh", stacked = True, figsize = (10,10), 
     title = "Proportion of customer segments by Bank Products");

plt.xlabel("Proportion of Customer Segments");

plt.legend(bbox_to_anchor = (0.7, -0.08), ncol = 4);
```

```python
temp_df.sort_values(by = [0,1,2,3]).\
plot(kind = "barh", stacked = True, figsize = (10,10), 
     title = "Proportion of customer segments by Bank Products");

plt.xlabel("Proportion of Customer Segments");

plt.legend(bbox_to_anchor = (0.7, -0.08), ncol = 4);
```
